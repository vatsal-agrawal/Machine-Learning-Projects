```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data = pd.get_dummies(columns=['Geography'],data=data)
```

```python
data = pd.get_dummies(columns=['Gender'],data=data,drop_first=True)
```

```python
data.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data.drop(['Exited'],axis=1),data.Exited,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
scaled_x_train = ss.fit_transform(x_train)
scaled_x_test = ss.transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
myclf = Sequential()
```

```python
myclf.add(Dense(activation='relu',output_dim=6,input_dim=12))
myclf.add(Dense(activation='relu',output_dim=6))
myclf.add(Dense(activation='sigmoid',output_dim=1))
myclf.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_func():
    myclf = Sequential()
    myclf.add(Dense(activation='relu',output_dim=12,input_dim=12))
    myclf.add(Dropout(0.1))
    myclf.add(Dense(activation='relu',output_dim=12))
    myclf.add(Dropout(0.1))
    myclf.add(Dense(activation='sigmoid',output_dim=1))
    myclf.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return myclf
```

```python
kclf = KerasClassifier(build_fn=build_func,batch_size=32,epochs=100)
```

```python
accuracy = cross_val_score(estimator=kclf,cv=10,n_jobs=-1,X=scaled_x_train,y=y_train)
```

```python
print(accuracy.mean())
print(accuracy.std())
```

```python
accuracy
```

```python
kclf.fit(scaled_x_train,y_train)
```

```python
yp = kclf.predict(scaled_x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,yp)
```
